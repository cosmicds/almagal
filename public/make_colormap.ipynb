From [wwt-webgl-engine#402](https://github.com/WorldWideTelescope/wwt-webgl-engine/pull/402)
```ts
colors = wwtlib.ColorMapContainer.viridis.colors.map((color, index) => [index, color.r, color.g, color.b]);
cm=wwtlib.ColorMapContainer.fromArgbList(colors); // also is a .fromStringList which takes hex

layer.get_imageSet().get_fitsProperties().colorMapName = "viridisalpha"
```


Need a `.ts` file with 
```ts
export colors = [
    [1, r1, g1, b1],
    [1, r2, g2, b2],
    ...
]

````python
# import matplotlib colormaps
from matplotlib import cm
from matplotlib.colors import rgb2hex
````

````python
COLORMAP = 'coolwarm'
cm.get_cmap(COLORMAP)
````

````python
COLORMAP = 'RdBu_r'
cm.get_cmap(COLORMAP)
````

````python
# need 256 colors
colors = cm.get_cmap(COLORMAP)(range(256))
hexColors = [rgb2hex(c) for c in colors]
# separate into sets of 8
hexColors = [hexColors[i:i+8] for i in range(0, len(hexColors), 8)]
with open(f'../src/assets/{COLORMAP}.ts', 'w') as f:
    f.write('export default [\n')
    for c in hexColors:
        line = ', '.join(f"'{color}'" for color in c)
        f.write(f'  {line},\n')
    f.write('];\n')
````
